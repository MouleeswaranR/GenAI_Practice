```python
from google.colab import userdata

api_key = userdata.get('OPENAI_API_KEY')
base_url=userdata.get('OPENROUTER_BASE_URL')
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool=DuckDuckGoSearchRun()

results=search_tool.invoke('top news in india today')

print(results)
```

```python
llm=ChatOpenAI(
    model="meta-llama/llama-3-8b-instruct",
    api_key=api_key,
    base_url=base_url,
)
```

```python
from langchain_classic.agents import AgentExecutor, create_react_agent
from langchain_classic import hub
```

```python
#pull the ReAct prompt

prompt=hub.pull("hwchase17/react")
```

```python
agent=create_react_agent(
    llm=llm,
    tools=[search_tool],
    prompt=prompt #like a system prompt for the agent
)
```

```python
#wrap it with agent executor

agent_executor=AgentExecutor(
    agent=agent,
    tools=[search_tool],
    verbose=True
)
```

```python
#invoke
result=agent_executor.invoke({"input":"what is the percentage of RCB winning ipl trophy in 2026 year"})
print(result)
```
